## Merging both files

```python
# Checking if 'ORIGIN' and 'DEST' fiels are having same unique data
dest_unique_values = df1['DEST'].unique()
origin_unique_values = df1['ORIGIN'].unique()

if all(dest_val in origin_unique_values for dest_val in dest_unique_values) and \
   all(origin_val in dest_unique_values for origin_val in origin_unique_values):
    print("The unique values in 'DEST' are the same as in 'ORIGIN'.")
else:
    print("The unique values in 'DEST' are different from 'ORIGIN'.")
```

#### Hence we can use either of the column for joining the files.

```python
# Merging both the files 
merged_df = pd.merge(df1, df, left_on='ORIGIN', right_on='airport', how='left')

```

```python
merged_df.head()
```

```python
# Drop redundant column
merged_df = merged_df.drop(['airport'], axis=1)
```
